# Compute : Delta(TOA), Delta(tas), Lambda

## The ensemble mean

```python
DeltaT_ensemble = np.load(file="DeltaT_NoOut_global_mean.npy")
DeltaT_ensemble_mean = DeltaT_ensemble.mean()
DeltaT_ensemble_mean
```

```python
nb_p_list = ['p105',  'p30',  'p65', 'p15', 'p109',  'p160',  'p34', 'p111', \
            'p122',  'p39', 'p123', 'p88', 'p184', 'p131',  'p49', \
            'p133',  'p193', 'p13',  'p195',  'p93', 'p140', 'p142',   'p58', \
            'p144', 'p148', 'p14', 'p209', 'p216', 'p218', 'p222', 'p227', 'p244', \
            'p245', 'p249', 'p254', 'p255', 'p262', 'p269', 'p279', 'p281', 'p287', 'p294', \
            'p297', 'p2', 'p4']

N=len(nb_p_list)
Lbda = [1]*N
Lbda_SW = [1]*N
Lbda_LW = [1]*N
Delta_T = [1]*N
i = 0
while i<N:
    nb_p = nb_p_list[i]
    
    amip_tas = amip[str(nb_p) + "_amip"]['tas']
    future4K_tas = future4K[str(nb_p) + "_future4K"]['tas']
    Delta_tas = future4K_tas - amip_tas
    Delta_T[i] = Delta_tas
    
    Delta_N = future4K_N[str(nb_p) + "_future4K"] - amip_N[str(nb_p) + "_amip"]
    Lambda = Delta_N/DeltaT_ensemble_mean
    Lbda[i] = Lambda
    
    Delta_SW = future4K_SW[str(nb_p) + "_future4K"] - amip_SW[str(nb_p) + "_amip"]
    Lambda_SW = Delta_SW/DeltaT_ensemble_mean
    Lbda_SW[i] = Lambda_SW
    
    Delta_LW = future4K_LW[str(nb_p) + "_future4K"] - amip_LW[str(nb_p) + "_amip"]
    Lambda_LW = Delta_LW/DeltaT_ensemble_mean
    Lbda_LW[i] = Lambda_LW
    
    i=i+1
```

```python
Lbda_matrix = xr.combine_nested(Lbda, concat_dim=['members'])
EM_Lambda = Lbda_matrix.mean(dim='members')
EM_Lambda
```

```python
LbdaSW_matrix = xr.combine_nested(Lbda_SW, concat_dim=['members'])
EM_Lambda_SW = LbdaSW_matrix.mean(dim='members')
EM_Lambda_SW
```

```python
LbdaLW_matrix = xr.combine_nested(Lbda_LW, concat_dim=['members'])
EM_Lambda_LW = LbdaLW_matrix.mean(dim='members')
EM_Lambda_LW
```

```python
DeltaT_matrix = xr.combine_nested(Delta_T, concat_dim=['members'])
EM_Delta_T = DeltaT_matrix.mean(dim='members')
EM_Delta_T
```
